```python
from pathlib import Path
import torch
import autorootcwd
from tqdm import tqdm
base_dir = Path("outputs/JiT_CondImg-H-16-OCTA500_6M")
pth_files = list(base_dir.rglob("*.pth"))

results = []
for pth in tqdm(pth_files):
    try:
        obj = torch.load(pth, weights_only=False, map_location="cpu")
        model = obj.get("model").keys()
        has_attr = "net.in_context_posemb" in model
        results.append((pth, has_attr))
        del obj["model"]["net.in_context_posemb"]
        torch.save(obj, pth)
    except Exception as e:
        results.append((pth, f"error: {e}"))

for path, status in results:
    print(path, status)
```
