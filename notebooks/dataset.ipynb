```python
import os
import autorootcwd
from matplotlib import pyplot as plt
from natsort import natsorted
import numpy as np

img_dir = "data/OCTA500_6M/val/images"
images = natsorted(os.listdir(img_dir))
label_dir = "data/OCTA500_6M/val/labels"
labels = natsorted(os.listdir(label_dir))
print(f"Number of images: {len(images)}")
for img_name, label_name in zip(images[:3], labels[:3]):
    img = os.path.join(img_dir, img_name)
    label = os.path.join(label_dir, label_name)
    img = plt.imread(img)
    label = plt.imread(label)
    print(f"Image: {img.shape}, {img.min()}, {img.max()}, {img.dtype}")
    print(f"Label: {label.shape}, {label.min()}, {label.max()}, {label.dtype}")
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].imshow(img)
    ax[0].set_title("Image")
    ax[0].axis('off')
    ax[1].imshow(label, cmap='gray')
    ax[1].set_title("Label")
    ax[1].axis('off')
    plt.show()
```

```python
import os
import autorootcwd
from matplotlib import pyplot as plt
from natsort import natsorted
import numpy as np

img_dir = "data/MoNuSeg/train/images"
images = natsorted(os.listdir(img_dir))
label_dir = "data/MoNuSeg/train/labels"
labels = natsorted(os.listdir(label_dir))
print(f"Number of images: {len(images)}")
for img_name, label_name in zip(images[:3], labels[:3]):
    img = os.path.join(img_dir, img_name)
    label = os.path.join(label_dir, label_name)
    img = plt.imread(img)
    label = np.load(label)
    print(f"Image: {img.shape}, {img.min()}, {img.max()}, {img.dtype}")
    print(f"Label: {label.shape}, {label.min()}, {label.max()}, {label.dtype}")
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].imshow(img)
    ax[0].set_title("Image")
    ax[0].axis('off')
    ax[1].imshow(label, cmap='gray')
    ax[1].set_title("Label")
    ax[1].axis('off')
    plt.show()
```

```python
import os
import autorootcwd
from matplotlib import pyplot as plt
import numpy as np
from natsort import natsorted
from collections import defaultdict

img_dir = "data/ISIC2016/test/images"
images = natsorted(os.listdir(img_dir))
label_dir = "data/ISIC2016/test/labels"
labels = natsorted(os.listdir(label_dir))

data = defaultdict(list)
print(f"Number of images: {len(images)}")
for img_name, label_name in zip(images, labels):
    img = os.path.join(img_dir, img_name)
    label = os.path.join(label_dir, label_name)
    img = plt.imread(img)
    label = plt.imread(label)
    data[img.shape[:2]].append((img_name, label_name))
for shape, items in data.items():
    print(f"Shape: {shape}, Number of images: {len(items)}")
```

```python
import os
import autorootcwd
from matplotlib import pyplot as plt
import numpy as np
from natsort import natsorted
from collections import defaultdict

img_dir = "data/ISIC2016/train/images"
images = natsorted([
    f for f in os.listdir(img_dir)
    if f.endswith('.jpg')
])
label_dir = "data/ISIC2016/train/labels"
labels = natsorted([
    f for f in os.listdir(label_dir)
    if f.endswith('.png')
])

data = defaultdict(list)
print(f"Number of images: {len(images)} / Number of labels: {len(labels)}")
for img_name, label_name in zip(images, labels):
    img = os.path.join(img_dir, img_name)
    label = os.path.join(label_dir, label_name)
    img = plt.imread(img)
    label = plt.imread(label)
    data[img.shape[:2]].append((img_name, label_name))
for shape, items in data.items():
    print(f"Shape: {shape}, Number of images: {len(items)}")
```
